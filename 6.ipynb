```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
```

```python
df=pd.read_csv("iris.csv")
df.head()
```

```python
x=df.drop(['Id','Species'],1)
y=df['Species']
x_train,y_train,x_test,y_test=train_test_split(x,y,test_size=0.2)
```

```python
fig,axs=plt.subplots(2,2)
axs[0,0].hist(x_train['sepallength'],bins=5)
axs[0,0].set_title("Sepal Length")
axs[0,1].hist(x_train['sepalwidth'],bins=5)
axs[0,1].set_title("Sepal Width")
axs[1,0].hist(x_train['petallength'],bins=5)
axs[1,0].set_title("Petal Length")
axs[1,1].hist(x_train['petalwidth'],bins=5)
axs[1,1].set_title("Petal Width")
fig.tight_layout(pad=1.0)
```

```python
fig,axs=plt.subplots(2,2)
cn=['Iris-setosa','Iris-versicolor','Iris-virginica']
sns.boxplot(x='Species',y='sepallength',data=df,order=cn,ax=axs[0,0]);
sns.boxplot(x='Species',y='sepalwidth',data=df,order=cn,ax=axs[0,1]);
sns.boxplot(x='Species',y='petallength',data=df,order=cn,ax=axs[1,0]);
sns.boxplot(x='Species',y='petalwidth',data=df,order=cn,ax=axs[1,1]);
fig.tight_layout(pad=1.0);
```

```python
fig,axs=plt.subplots(2,2)
sns.violinplot(x='Species',y='sepallength',data=df,order=cn,ax=axs[0,0]);
sns.violinplot(x='Species',y='sepalwidth',data=df,order=cn,ax=axs[0,1]);
sns.violinplot(x='Species',y='petallength',data=df,order=cn,ax=axs[1,0]);
sns.violinplot(x='Species',y='petalwidth',data=df,order=cn,ax=axs[1,1]);
fig.tight_layout(pad=1.0);
```

```python
corr=df.corr()
sns.heatmap(corr,annot=True,square=True)
```

```python
plt.figure(figsize=(10,5))
x=sns.countplot(x='sepallength',hue='Species',data=df,palette='rainbow')
x.legend(bbox_to_anchor=(1,1),loc='upper left')
```
